```python


##.Créer un tableau de 4 lignes et 3 colonnes rempli de 11 nombres entiers aléatoires compris entre 0 et 9
Tab = np.random.randint(0,10
,(4,3))
Tab
```

```python

##le type des éléments du tableau
Tab.dtype
```

```python


## changer le type des éléments du tableau en float32
Tab=np.float32(Tab)
Tab
```

```python

## ajouter 3 à la fin du tableau à une dimension
Tab=np.append(Tab,3)
Tab
```

```python

##Transformer le tableau en matrice dénommée “a” de 3 lignes et  4colonnes
a= Tab
a=np.delete(a,-1)
a=a.reshape(3,4)
a
```

```python

## Créer une matrice dénommée“b” de 3 lignes et 4 colonnes remplie de série d'entiers successifs allant de 1 à 12 inclus
b= np.arange(1,13)
b=b.reshape(3,4)
b
```

```python

## Créer une matrice dénommée“c” qui est la concaténation des deux tableaux “a” et“b” dans le sens horizontale
c= np.hstack((a,b))
c
```

```python

## Créer une matrice dénommée“d” qui est la concaténation des deux tableaux “a” et“b” dans le sens verticale
d=  np.vstack((a,b))
d
```

```python

##aplatir le tableau d
d= np.ravel(d)
d
```

```python

##supprimer le dernier élément dans le tableau d
d=np.delete(d,-1)
d
```
